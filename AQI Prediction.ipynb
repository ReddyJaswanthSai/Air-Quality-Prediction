```python
pm_ = list(df['PM 2.5'])

Acceptable = []

for pm in pm_:
    if pm >= 151:
        Acceptable.append(0)
    else:
        Acceptable.append(1)

df['Acceptable'] = Acceptable

df.sample(3)
```

```python
cols =  list(df.columns)

plt.figure(figsize=(20, 20))

for i in range(1, 9):
    plt.subplot(3, 4, i)
    sns.scatterplot(x = cols[i - 1], y = df['PM 2.5'],data = df, hue = "Acceptable", palette = "husl")
```

```python
corrmat = df.corr()
top_corr_features = corrmat.index[:-1] # dropping 'Acceptable'

plt.figure(figsize=(10,10))

#plot heat map
sns.heatmap(df[top_corr_features].corr(),annot = True, linewidths=.5)
# print(corrmat)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df[['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM','CO2','CH4']]
y = df['PM 2.5']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.10, random_state = 42)
```

```python
print('Length of X_train', len(X_train))
print('Length of X_val', len(X_val))
print('Length of y_train', len(y_train))
print('Length of y_val', len(y_val))
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
classifier = RandomForestRegressor(n_estimators = 300, random_state = 0) #N-estimators means how many time predict the value like that 
classifier.fit(X_train, y_train)

print("classifier .score : ",classifier  .score(X_val, y_val))
```

```python
print("classifier .score of training data: ", classifier .score(X_train, y_train))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(classifier  , X, y, cv = 5)

print(score)
print("Mean score : ",score.mean())
```

# support vector machine

```python
from sklearn.svm import SVR
regressor1 = SVR()
regressor1.fit(X_train, y_train)

print("regressor1 .score : ",regressor1  .score(X_val, y_val))
```

```python
print("regressor .score of training data: ", regressor1.score(X_train, y_train))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(regressor1  , X, y, cv = 5)

print(score)
print("Mean score : ",score.mean())
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

ram = LinearRegression()
ram.fit(X_train, y_train)


print("ram .score : ", ram .score(X_val, y_val))
print("ram .coef_ : ",ram .coef_)
print("ram .intercept_ : ",ram .intercept_)
```

```python
print("ram .score of training data: ", ram .score(X_train, y_train))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(ram  , X, y, cv = 5)

print(score)
print("Mean score : ",score.mean())
```

```python
coeff_df = pd.DataFrame(ram .coef_, X.columns, columns = ['Coefficient'])
coeff_df
```

```python
from sklearn import metrics

prediction = classifier .predict(X_val)

print('MAE:', metrics.mean_absolute_error(y_val, prediction))
print('MSE:', metrics.mean_squared_error(y_val, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_val, prediction)))
```
